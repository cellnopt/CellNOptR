```python
%load_ext rmagic
```

```python
%R library(CellNOptR)
```

```python
%R cnolist = CNOlist(CNOdata("MD-ToyMMB_T2.csv"))
```

```python
%R pknmodel = readSIF(CNOdata('PKN-ToyMMB_T2.sif'))
```

```python
%R model = preprocessing(cnolist, pknmodel, compression=TRUE)
```

```python
%R optT1 = gaBinaryT1(cnolist, model, verbose=FALSE)
%R optTN = gaBinaryTN(cnolist, model, verbose=FALSE, bStrings=list(optT1$bString))
```

```python
%R res = cutAndPlot(cnolist, model, bStrings=list(optT1$bString))
```

The second optimisation results can be shown using the same function but providing the two optimisation bitstrings:

```python
%R res2 = cutAndPlot(cnolist, model, bStrings=list(optT1$bString, optTN$bString))
```
