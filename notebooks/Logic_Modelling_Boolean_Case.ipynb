<h1>Example of boolean logic modelling with CellNOptR</h1>

- reading a model (ToyMMB.sif) in SIF format
- reading a dats set (ToyMMB.csv) in MIDAS format
- preprocess the data
- performing a genetic algorithm optimisation
- looking at the results

```python
%load_ext rmagic
%R library(CellNOptR)
```

```python
%%R 
pknmodel = readSIF(CNOdata("PKN-ToyMMB.sif"))
cnolist = CNOlist(CNOdata("MD-ToyMMB.csv"))
```

```python
%%R
model = preprocessing(cnolist, pknmodel)
```

```python
%%R 
plotModel(model, cnolist)
```

```python
%%R 
plot(cnolist)
```

```python
%%R
 # This may take time depending on your problem
 # various parameters can be used to tuned the optimisation.
 # see www.cellnopt.org documentation.
 res = gaBinaryT1(cnolist,model, verbose=F)
```
